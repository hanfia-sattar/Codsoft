```python
from tkinter import *

root = Tk()
# Set geometry(withxheight)
root.geometry('360x135')
root.resizable(False, False)
root.title('Calculator')
root.configure(background="black")

calculate = ""
operation = StringVar()

e1 = Entry(root, textvariable=operation)
e1.grid(columnspan=7, ipadx=120, ipady=10)


def clicked(val):
    global calculate
    calculate = calculate + str(val)
    operation.set(calculate)


def clear():
    global calculate
    calculate = ""
    operation.set("")


def perform_operation():
    try:
        global calculate
        total = str(eval(calculate))
        operation.set(total)
        calculate = ""
    except:
        operation.set(" error ")
        calculate = ""


button1 = Button(root, text=' 1 ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                 height=1, width=7, command=lambda: clicked(1))

button1.grid(row=2, column=0)

button2 = Button(root, text=' 2 ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                 height=1, width=7, command=lambda: clicked(2))
button2.grid(row=2, column=1)

button3 = Button(root, text=' 3 ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                 height=1, width=7, command=lambda: clicked(3))
button3.grid(row=2, column=2)

button4 = Button(root, text=' 4 ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                 height=1, width=7, command=lambda: clicked(4))
button4.grid(row=2, column=3)

button_multi = Button(root, text=' x ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                      height=1, width=7, command=lambda: clicked('*'))
button_multi.grid(row=2, column=4)

# new row
button5 = Button(root, text=' 5 ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                 height=1, width=7, command=lambda: clicked(5))
button5.grid(row=3, column=0)

button6 = Button(root, text='6 ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                 height=1, width=7, command=lambda: clicked(6))
button6.grid(row=3, column=1)

button7 = Button(root, text=' 7 ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                 height=1, width=7, command=lambda: clicked(7))
button7.grid(row=3, column=2)

button8 = Button(root, text=' 8 ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                 height=1, width=7, command=lambda: clicked(8))
button8.grid(row=3, column=3)

button_add = Button(root, text=' + ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                    height=1, width=7, command=lambda: clicked('+'))
button_add.grid(row=3, column=4)

# row 3
button9 = Button(root, text=' 9 ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                 height=1, width=7, command=lambda: clicked(9))
button9.grid(row=4, column=0)

button0 = Button(root, text=' 0 ', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                 height=1, width=7, command=lambda: clicked(0))
button0.grid(row=4, column=1)

button_sub = Button(root, text=' -', fg='white', bg='blue', font=('Sans', '11', 'bold'),
                    height=1, width=7, command=lambda: clicked('-'))
button_sub.grid(row=4, column=2)

button_del = Button(root, text=' C ', fg='Blue', bg='white', font=('Sans', '11', 'bold'),
                    height=1, width=7, command=clear)
button_del.grid(row=4, column=3)

button_eql = Button(root, text=' = ', fg='Blue', bg='white', font=('Sans', '11', 'bold'),
                    height=1, width=7, command=perform_operation)
button_eql.grid(row=4, column=4)

root.mainloop()
```
